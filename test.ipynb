```python
import uiautomation as auto
import time
import pyautogui
import re
import pinyin

time.sleep(3)

root = auto.GetRootControl()
taskbar = None
for control in root.GetChildren():
    if control.Name == '任务栏':
        taskbar = control
        break
if taskbar is None:
    print('没有找到任务栏，退出')
    quit()

input_indicator = None
for child_depth2 in taskbar.GetChildren():
    for child_depth3 in child_depth2.GetChildren():
        for child_depth4 in child_depth3.GetChildren():
            child_depth4_name = child_depth4.Name
            match = re.search('托盘输入指示器', child_depth4.Name)
            if match:
                input_indicator = child_depth4
                break

if input_indicator is None:
    print('没有找到托盘输入指示器，退出')
    quit()

# 切换到微软拼音
if re.search('英语\(', input_indicator.Name):
    # Windows + 空格
    pyautogui.hotkey('win', 'space')
time.sleep(0.5)
# 切换到中文模式
if re.search('英语模式', input_indicator.Name):
    # shift
    pyautogui.hotkey('shift')
time.sleep(0.5)

print(input_indicator.Name)

# 模拟键盘输入

input_experience = None
candidate_ui = None
candidate_panel = None

ui_respond_time = 0.06
type_interval = 0.001

def pinyin_input(charater):
    global input_experience
    global candidate_ui
    global candidate_panel

    if character == '，':
        pyautogui.press(',')
        return
    elif character == '。':
        pyautogui.press('.')
        return
    elif character == '？':
        pyautogui.press('?')
        return
    elif character == '！':
        pyautogui.press('!')
        return
    elif character == '、':
        pyautogui.press('\\')
        return
    elif character == '；':
        pyautogui.press(';')
        return
    elif character == '：':
        pyautogui.press(':')
        return
    elif character == '“':
        pyautogui.press('"')
        return
    elif character == '”':
        pyautogui.press('"')
        return
    elif character == '‘':
        pyautogui.press('\'')
        return
    elif character == '’':
        pyautogui.press('\'')
        return
    elif character == '《':
        pyautogui.press('<')
        return
    elif character == '》':
        pyautogui.press('>')
        return
    elif character == '（':
        pyautogui.press('(')
        return
    elif character == '）':
        pyautogui.press(')')
        return
    elif character == '【':
        pyautogui.press('[')
        return
    elif character == '】':
        pyautogui.press(']')
        return
    elif character == '——':
        pyautogui.press('-') # 待修改
        return
    elif character == '…':
        pyautogui.press('...') # 待修改
        return
    elif character == '·':
        pyautogui.press('`')
        return

    character_pinyin = pinyin.get(character, format="strip")
    # print(character_pinyin)
    pyautogui.typewrite(character_pinyin, interval=type_interval)
    time.sleep(ui_respond_time)

    if candidate_panel is None:
        input_experience = auto.WindowControl(searchDepth=2, Name='Windows 输入体验')
        print(input_experience)
        candidate_ui = input_experience.MenuControl(searchDepth=1, Name='Microsoft 候选项 UI')
        candidate_panel = candidate_ui.ListControl(searchDepth=1, Name='候选项面板')
    
    candidates = candidate_panel.GetChildren()

    hit = False

    while not hit:
        candidate_num = 0

        for candidate in candidates:
            if candidate.ControlType != auto.ControlType.ListItemControl:
                continue
            candidate_num += 1
            if candidate.Name == character:
                pyautogui.press(str(candidate_num))
                hit = True
                break
        
        if not hit:
            pyautogui.press(']')
            time.sleep(ui_respond_time)


characters = "亲爱的朋友，面对压力和焦虑，首先要做的是调整自己的心态。试着将注意力转移，进行一些放松的活动，比如散步、听音乐、做瑜伽等等。此外，可以寻求朋友和家人的帮助，与他们分享自己的感受。如果情况严重，也可以考虑寻求专业的心理咨询师的帮助。请记住，你并不孤单，许多人都曾经面临过类似的挑战，你并不孤单，有人会愿意帮助你。加油！"

import datetime

print('start:')
print(datetime.datetime.now())

for character in characters:
    pinyin_input(character)
    time.sleep(ui_respond_time)

print('end:')
print(datetime.datetime.now())

print(f'character count: {len(characters)}')

亲爱的朋友，面对压力和焦虑，首先要做的是调整自己的心态。试着将注意力转移，进行一些放松的活动，比如散步、听音乐、做瑜伽等等。此外，可以寻求朋友和家人的帮助，与他们分享自己的感受。如果情况严重，也可以考虑寻求专业的心理咨询师的帮助。请记住，你并不孤单，许多人都曾经面临过类似的挑战，你并不孤单，有人会愿意帮助你。加油！
```

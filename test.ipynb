```python
import uiautomation as auto
import time
import pyautogui
import re
import pinyin

time.sleep(3)

root = auto.GetRootControl()
taskbar = None
for control in root.GetChildren():
    if control.Name == '任务栏':
        taskbar = control
        break
if taskbar is None:
    print('没有找到任务栏，退出')
    quit()

input_indicator = None
for child_depth2 in taskbar.GetChildren():
    for child_depth3 in child_depth2.GetChildren():
        for child_depth4 in child_depth3.GetChildren():
            child_depth4_name = child_depth4.Name
            match = re.search('托盘输入指示器', child_depth4.Name)
            if match:
                input_indicator = child_depth4
                break

if input_indicator is None:
    print('没有找到托盘输入指示器，退出')
    quit()

# 切换到微软拼音
if re.search('英语\(', input_indicator.Name):
    # Windows + 空格
    pyautogui.hotkey('win', 'space')
while re.search('英语\(', input_indicator.Name):
    pass
# 切换到中文模式
if re.search('英语模式', input_indicator.Name):
    # shift
    pyautogui.hotkey('shift')
while re.search('英语模式', input_indicator.Name):
    pass

print(input_indicator.Name)

# 模拟键盘输入

input_experience = None
candidate_ui = None
candidate_panel = None
previous_page_button = None
next_page_button = None

ui_respond_time = 0.06
type_interval = 0.001

def pinyin_input(charater):
    global input_experience
    global candidate_ui
    global candidate_panel
    global previous_page_button
    global next_page_button

    if character == '，':
        pyautogui.press(',')
        return
    elif character == '。':
        pyautogui.press('.')
        return
    elif character == '？':
        pyautogui.press('?')
        return
    elif character == '！':
        pyautogui.press('!')
        return
    elif character == '、':
        pyautogui.press('\\')
        return
    elif character == '；':
        pyautogui.press(';')
        return
    elif character == '：':
        pyautogui.press(':')
        return
    elif character == '“':
        pyautogui.press('"')
        return
    elif character == '”':
        pyautogui.press('"')
        return
    elif character == '‘':
        pyautogui.press('\'')
        return
    elif character == '’':
        pyautogui.press('\'')
        return
    elif character == '《':
        pyautogui.press('<')
        return
    elif character == '》':
        pyautogui.press('>')
        return
    elif character == '（':
        pyautogui.press('(')
        return
    elif character == '）':
        pyautogui.press(')')
        return
    elif character == '【':
        pyautogui.press('[')
        return
    elif character == '】':
        pyautogui.press(']')
        return
    elif character == '——':
        pyautogui.press('-') # 待修改
        return
    elif character == '…':
        pyautogui.press('...') # 待修改
        return
    elif character == '·':
        pyautogui.press('`')
        return

    character_pinyin = pinyin.get(character, format="strip")
    # print(character_pinyin)
    pyautogui.typewrite(character_pinyin, interval=type_interval)
    time.sleep(ui_respond_time)

    if candidate_panel is None:
        input_experience = auto.WindowControl(searchDepth=2, Name='Windows 输入体验')
        print(input_experience)
        candidate_ui = input_experience.MenuControl(searchDepth=1, Name='Microsoft 候选项 UI')
        candidate_panel = candidate_ui.ListControl(searchDepth=1, Name='候选项面板')
        previous_page_button = candidate_panel.ButtonControl(searchDepth=1, Name='上一页')
        next_page_button = candidate_panel.ButtonControl(searchDepth=1, Name='下一页')
    
    candidates = candidate_panel.GetChildren()

    print(f'pre: {previous_page_button.IsEnabled}, next: {next_page_button.IsEnabled}')

    hit = False

    while not hit:
        candidate_num = 0

        for candidate in candidates:
            if candidate.ControlType != auto.ControlType.ListItemControl:
                continue
            candidate_num += 1
            if candidate.Name == character:
                pyautogui.press(str(candidate_num))
                hit = True
                break
        
        if not hit:
            pyautogui.press(']')
            time.sleep(ui_respond_time)


characters = "噫吁嚱，危乎高哉！蜀道之难，难于上青天！蚕丛及鱼凫，开国何茫然！尔来四万八千岁，不与秦塞通人烟。西当太白有鸟道，可以横绝峨眉巅。地崩山摧壮士死，然后天梯石栈相钩连。上有六龙回日之高标，下有冲波逆折之回川。黄鹤之飞尚不得过，猿猱欲度愁攀援。青泥何盘盘，百步九折萦岩峦。扪参历井仰胁息，以手抚膺坐长叹。问君西游何时还？畏途巉岩不可攀。但见悲鸟号古木，雄飞雌从绕林间。又闻子规啼夜月，愁空山。蜀道之难，难于上青天，使人听此凋朱颜！连峰去天不盈尺，枯松倒挂倚绝壁。飞湍瀑流争喧豗，砯崖转石万壑雷。其险也如此，嗟尔远道之人胡为乎来哉！剑阁峥嵘而崔嵬，一夫当关，万夫莫开。所守或匪亲，化为狼与豺。朝避猛虎，夕避长蛇，磨牙吮血，杀人如麻。锦城虽云乐，不如早还家。蜀道之难，难于上青天，侧身西望长咨嗟！"

import datetime

print('start:')
print(datetime.datetime.now())

for character in characters:
    pinyin_input(character)
    time.sleep(ui_respond_time)

print('end:')
print(datetime.datetime.now())

print(f'character count: {len(characters)}')
```
